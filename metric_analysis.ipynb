# Metric Analysis
Plot training losses

```python
import matplotlib.pyplot as plt
import json
import numpy as np
import os
```

```python
model_name = 'Custom Backbone Full Dataset 100k-260k'
metrics_path = '/content/metrics.json'
```

```python
with open(metrics_path, 'r') as file:
    content = file.read()
    json_objects = content.split('\n')

i = 0
data_lst = []
for obj in json_objects[:-1]:
  data_lst.append(json.loads(obj))
data_lst[0]
```

```python
total_loss = []
itr = []
for data in data_lst:
  if 'total_loss' in data.keys():
    if len(itr) > 0 and data['iteration'] < itr[-1]:
      total_loss = []
      itr = []
    total_loss.append(data['total_loss'])
    itr.append(data['iteration'])
```

```python
plt.plot(itr, total_loss)

plt.title(f'{model_name} Training Loss')
plt.xlabel('Iterations')
plt.ylabel('Total Loss')
plt.ylim(-.3, 8.3)

plt.savefig('/content/loss_graph.png', dpi=300)
plt.show()
```
